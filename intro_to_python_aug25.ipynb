```python
def year_to_csv(data, year):
  '''
  filter data for a year and save into csv file
  data: a dataframe containing range of year
  year: year to be filtered out
  '''
  year_data = data[data.year == year]
  # save this to csv
  filename = 'surveys_' + str(year) + ".csv"
  year_data.to_csv(filename, index = False)
```

```python
year_to_csv(year=2005, data = all_data)
```
